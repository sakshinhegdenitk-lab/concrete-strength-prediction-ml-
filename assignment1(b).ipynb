```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Traffic Load":["High", "High", "Medium", "Medium" , "Low", "Low"],
    "Temperature":[40,35,27.5,28,20,25],
    "Rainfall":[125,120,90,80,45,50],
    "Pavement Life":[8,7,13,12,17,15]
}
```

```python
# Features:
# [traffic load, temperature(celsius),rainfall(mm)]

df=pd.DataFrame(data)

traffic_mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}

df["Traffic Load"] = df["Traffic Load"].map(traffic_mapping)

print("\nDataset:")
print(df)
X = np.array([
    [3, 40, 125],
    [3, 35, 120],
    [2, 27.5, 90],
    [2, 28, 80],
    [1, 20, 45],
    [1, 25, 50]
])

# Target Output:
# pavement life

y = np.array([8,7,13,12,17,15])
```

```python
X = df[["Traffic Load", "Temperature", "Rainfall"]]

y = df["Pavement Life"]
```

```python
plt.figure(figsize=(8,6))

# Scatter plot 1
plt.scatter(df['Traffic Load'], df['Pavement Life'])

plt.xlabel("Traffic load")
plt.ylabel("Pavement life")

plt.title("Traffic load vs Pavement life")

plt.grid(True)
plt.xticks([1, 2, 3], ['Low', 'Medium', 'High'])

plt.show()

```

```python
plt.figure(figsize=(8,6))

# Scatter plot 2
plt.scatter(df['Temperature'], df['Pavement Life'])

plt.xlabel("Temperature")
plt.ylabel("Pavement life")

plt.title("Temperature vs Pavement life")

plt.grid(True)


plt.show()
```

```python
plt.figure(figsize=(8,6))
# scatter plot3
plt.scatter(df['Rainfall'], df['Pavement Life'])

plt.xlabel("Rainfall")
plt.ylabel("Pavement Life ")
plt.title("Rainfall vs Pavement Life")
plt.grid(True)

plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)

print("Model trained successfully!")
```

```python
#Coefficients (weights)

print("Traffic load coefficient:", model.coef_[0])

print("Temperature coefficient :", model.coef_[1])

print("rainfall coefficient  :", model.coef_[2])

# Intercept

print("\nIntercept :", model.intercept_)
```

```python
y_pred = model.predict(X)
print("Predicted Life:\n")
print(y_pred)
```

```python
# 3D plot

fig = plt.figure(figsize=(10,7))

ax = fig.add_subplot(111, projection='3d')

# Scatter plot

scatter = ax.scatter(
    df['Traffic Load'],
    df['Temperature'],
    df['Rainfall'],
    c=df['Pavement Life']
)

# Labels

ax.set_xlabel("traffic load")
ax.set_ylabel("temperature")
ax.set_zlabel("rainfall")
plt.colorbar(scatter, label='Pavement Life (Years)')
plt.title('3D Scatter Plot of Pavement Life Data')
plt.show()
```

```python
#prediction of pavement life for a new road condition
new_data = pd.DataFrame({
    'Traffic_Load': [2, 3],
    'Temperature': [32, 38],
    'Rainfall': [130, 170]
})

predictions = model.predict(new_data)
print(f"Prediction for condition 1:, {predictions[0]:.2f}Years")
print(f"Prediction for condition 2:, {predictions[1]:.2f}Years")
```
