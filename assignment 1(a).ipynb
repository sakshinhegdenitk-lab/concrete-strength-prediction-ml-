```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#speed(km/hr)
X=np.array([40,60,80,100,140,160,180,200]).reshape(-1,1)
```

```python
#fuel consumption(L/hr)
y=np.array([3,4,5.5,7,13,14,17,20])
```

```python
print("speed:\n",X)
print("\nFuel Conspution:\n",y)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(X, y, s=100)

plt.xlabel("speed(km/hr)")
plt.ylabel("fuel consumption(L/hr)")

plt.title("speed vs fuel consumption")

plt.grid(True)

plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)

print("Model trained successfully!")
```

```python
# Slope
m = model.coef_[0]

# Intercept
b = model.intercept_

print("Slope =", m)
print("Intercept =", b)
```

```python
y_pred = model.predict(X)

print("Predicted Prices:\n")
print(y_pred)
```

```python
plt.figure(figsize=(8,6))

# Actual data
plt.scatter(X, y, s=100, label="Actual speed")

# Regression line
plt.plot(X, y_pred, linewidth=3, label="Regression Line")

plt.xlabel("speed(km/hr)")
plt.ylabel("fuel consumption(L/hr)")

plt.title("Linear Regression")

plt.legend()
plt.grid(True)

plt.show()
```

```python
#prediction of fuel consumption for a speed of 120km/hr
new_size = [[120]]

predicted_price = model.predict(new_size)

print(f"Predicted fuel cossuption for 120 km/hr = {predicted_price[0]:.2f} L/hr")
```

```python
#plotting residual errors
residuals = y - y_pred

print("\nResidual Errors:\n")

print(residuals)

plt.scatter(y_pred, residuals)

plt.axhline(y=0)

plt.xlabel("Predicted Values")
plt.ylabel("Residual Errors")
plt.title("Residual Plot")

plt.show()
```
